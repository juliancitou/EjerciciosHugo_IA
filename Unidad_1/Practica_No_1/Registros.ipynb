## PRACTICA DE REGISTROS (Practica n1)

### 1. CREACION DE LA LISTA TECNOLOGICO QUE ALMACENARA LOS DATOS

```python
# Lista global que funcionará como nuestra "base de datos"
tecnologicos = []
```

### 2. OPCION 1: Añadir tecnologico 

```python
def añadir_tecnologico():
    """Pide datos al usuario y los guarda en un diccionario."""
    print("\n--- Añadir Nuevo Tecnológico ---")
    campus = input("Nombre del Campus: ")
    estado = input("Estado: ")
    municipio = input("Municipio: ")
    # Convertimos la entrada a booleano de forma sencilla
    es_federal = input("¿Es federal? (si/no): ").lower() == 'si'
    
    # Creamos el diccionario con la estructura solicitada
    nuevo_reg = {
        "campus": campus,
        "estado": estado,
        "municipio": municipio,
        "federal": es_federal
    }
    
    tecnologicos.append(nuevo_reg)
    print("¡Registro añadido con éxito!")
```

### 3. OPCION 2: Borrar tecnologico existente

```python
def borrar_tecnologico():
    """Busca por nombre de campus y lo elimina de la lista."""
    nombre = input("\nNombre del campus a eliminar: ")
    global tecnologicos
    # Filtramos la lista: mantenemos todo lo que NO coincida con el nombre
    longitud_inicial = len(tecnologicos)
    tecnologicos = [t for t in tecnologicos if t['campus'].lower() != nombre.lower()]
    
    if len(tecnologicos) < longitud_inicial:
        print(f"El campus '{nombre}' ha sido eliminado.")
    else:
        print("No se encontró ningún campus con ese nombre.")
```

### 4. OPCION 3: Buscar tecnologico existente 

```python
def buscar_tecnologico():
    """Busca un campus específico y muestra su información."""
    nombre = input("\nNombre del campus a consultar: ")
    encontrado = False
    for t in tecnologicos:
        if t['campus'].lower() == nombre.lower():
            print(f"\nResultado: {t}")
            encontrado = True
            break
    if not encontrado:
        print("Tecnológico no encontrado.")
```

### 5. OPCION 4: Mostrar una lista de los tecnologicos registrados

```python
def listar_tecnologicos():
    """Muestra todos los registros almacenados."""
    print("\n--- Lista de Tecnológicos ---")
    if not tecnologicos:
        print("La lista está vacía.")
    else:
        for i, t in enumerate(tecnologicos, 1):
            print(f"{i}. {t['campus']} - {t['estado']} (Federal: {t['federal']})\n\n\n")
```

### 6. CREACIÓN DE MENU INTERACTIVO

```python
def menu():
    """Función principal que controla el flujo del programa."""
    while True:
        print("\n" + "="*20)
        print("   TECNOLÓGICOS")
        print("="*20)
        print("1. Añadir tecnologico")
        print("2. Borrar tecnologico")
        print("3. Buscar tecnologico")
        print("4. Listar tecnologicos")
        print("5. Salir")
        
        opcion = input("\nSelecciona una opción: ")
        
        if opcion == '1':
            añadir_tecnologico()
        elif opcion == '2':
            borrar_tecnologico()
        elif opcion == '3':
            buscar_tecnologico()
        elif opcion == '4':
            listar_tecnologicos()
        elif opcion == '5':
            print("Saliendo del programa...")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Ejecución del programa
menu()
```
